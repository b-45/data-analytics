**A stacker is a large machine used in bulk material handling. Its function is to pile bulk material such as limestone, ores and cereals on to a stockpile.**
***
A coal company in Western Australia requires some analytical work.  They operate 5 machines that stacks and reclaims mined coals 24-hours/ 365 days.  They can't afford to let these machines run idle or else they come up short by millions in revenue. 

After meeting with engineers and stakeholders, brainstorming, gathering requirements, a working plan to tackle the problem was conceived.

The challenge is to use the metric **idle capacity** to asesss which of the machines require maintenance.
These machines require maintenance when within the **previous month** there was at least one **8-hour period** when the average idle was over **10%**.

The idle capacity metric is defined as follows:




$ Idle\,Capacity =  {(Actual\,Tonnes - Nominal\,Capacity) \over Nominal\,Capacity} $

Idle capacity represents unused capacity for the hour.
In a nutshell, the tasks requires us to assess if the average of 8 hours exceeds 10% then the machine requires 
maintenance.


#### Some info about the data:

Datetime - hourly.

Actual Tonnage - utilization per hour.

Nominal Capacity - Nominal capacity for the machine is 4200 tons. Think of nominal as the approximation of accepted conditions.



We will break this process down into 3 steps.

1. Inspect and clean data

2. Compute metric

3. Communicate our results.

### Clean and Inspect data

```python
#import library to structure data
import pandas as pd 
# import library to make plots with custom settings
import matplotlib.pyplot as plt
import matplotlib.pyplot as mpl
%matplotlib inline
plt.rcParams['figure.figsize'] = (15, 5)
#mpl.style.use('ggplot')
```

```python
# load data into a pandas dataframe, parse the dates and set 'Datetime' as index.
data = pd.read_csv('rl1.csv', parse_dates=True, index_col='Datetime')
```

```python
#mpl.style.available
```

```python
# inspect table
data.head(3)
```

```python
# 720 rows of data
data.info()
```

```python
#  keep only the  columns we need.
#data = data[['Nominal Capacity', 'Actual Tonnes']]
```

```python
# rename columns for ease of use
data = data.rename(columns={'Nominal Capacity': 'nominal_capacity', 'Actual Tonnes':'actual_tonnes'})
```

```python
# inspect table
data.head()
```

***
### Compute Metric




$ Idle\,capacity = {(Actual\,Tonnes - Nominal\,Capacity) \over Nominal\,Capacity} $


```python
# calculate idle capacity metric
data['idle_capacity'] = (data['actual_tonnes'] - data['nominal_capacity'])/data['nominal_capacity']
```

```python
# calculate moving average with a 8-hour window
data['moving_average'] = data.idle_capacity.rolling(window=8,center=False).mean()
```

```python
# inspect the first 15 rows of the table using head function
data.head(15)
```

```python
# display plots to gauge the data

plt.subplot(311)
plt.plot(data['actual_tonnes'], label='Actual Tonnes')
plt.legend(loc='best')


plt.subplot(312)
plt.plot(data['idle_capacity'], label='Idle Capacity', color='crimson')
plt.legend(loc='best')

plt.subplot(313)
plt.plot(data['moving_average'],label='Moving Average', color='green')
plt.legend(loc='best')

plt.tight_layout()
```

```python
# plot the 8-hour moving average with a 10% reference line. plot looks better if we invert by mulitplying by -1.
plt.plot(data['moving_average'] * -1)
plt.axhline(y=.10, color='r')
plt.show()
```

```python
# filter points that exceeds 10% threshold
spikes = data[data['moving_average']  < -0.10]
```

```python
# display hours that exceeds threshold.
spikes
```

```python
# plot days that exceed the 10% threshold.
spikes.moving_average.resample('D').count().plot(kind='bar', )
```
